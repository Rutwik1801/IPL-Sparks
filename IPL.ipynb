# batsman runs contribute majorly to the total runs,while wide byes legbyes all contribute to extra runs

```python
batsman=deliveries.groupby(['match_id','inning','batting_team','batsman'])
batsmen=batsman['batsman_runs'].sum().reset_index()
batsmen=matches[['id','season']].merge(batsmen,left_on='id',right_on='match_id',how='left').drop('id',axis=1)
```

```python
batsmen
```

```python
runs_per_season=batsmen.groupby(['season','batting_team','batsman'])['batsman_runs'].sum().reset_index()
runs_per_season=runs_per_season.groupby(['season','batsman'])['batsman_runs'].sum().unstack().T
runs_per_season['total']=runs_per_season.sum(axis=1)
runs_per_season=runs_per_season.sort_values(by='total',ascending=False).drop('total',axis=1)
```

```python
runs_per_season
```

```python
plt.figure(figsize=(14,12))
runs_per_season[:5].T.plot()
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

## Top 5 highest run-scoring batsmen throughout IPL seasons

```python
bowler=deliveries.groupby(['match_id','inning','bowling_team','player_dismissed','bowler'])
bowlers=bowler['batsman_runs'].sum().reset_index().drop('batsman_runs',axis=1)
bowlers['wicket']=1
bowlers=matches[['id','season']].merge(bowlers,left_on='id',right_on='match_id',how='left').drop('id',axis=1)
```

```python
bowlers.head(5)
```

```python
wickets_per_season=bowlers[['season','bowler','wicket']]
wickets_per_season=wickets_per_season.groupby(['season','bowler'])['wicket'].sum().reset_index()
wickets_per_season=wickets_per_season.groupby(['season','bowler'])['wicket'].sum().unstack().T
wickets_per_season['total']=wickets_per_season.sum(axis=1)
wickets_per_season=wickets_per_season.sort_values(by='total',ascending=False)
```

```python
wickets_per_season.fillna(0,axis=1)
highest_wicket_takers=wickets_per_season.drop('total',axis=1)
```

```python
highest_wicket_takers[:5].T.plot()
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
